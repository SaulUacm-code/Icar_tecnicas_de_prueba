# <span style="color:#0485CF"> Estructura de Datos: Diccionarios (`dict`)</span>

#### A diferencia de las listas que se indexan por posición (0, 1, 2...), los diccionarios utilizan **Claves** (keys) únicas para acceder a sus **Valores** (values).

* #### **Clave-Valor**: Cada elemento es un par `llave: valor`.
* #### **Dinámicos**: Puedes agregar o quitar elementos en cualquier momento.
* #### **Indexados por Clave**: No importa el orden, accedes directamente por el nombre de la llave.
* #### **Anidables**: El valor de una llave puede ser otro diccionario, una lista o cualquier otro tipo.

> **Analogía**: Imagina una guía telefónica donde el **Nombre** es la clave y el **Número** es el valor. No buscas por la página 5, buscas por "Juan Pérez".

### Diccionario inicializado

````python
# 1. Creación de un diccionario (Perfil de Usuario de Prueba)

usuario: dict[str, int | str | bool] = {
        "id": 101,
        "username": "qa_tester_01",
        "email": "tester@empresa.com",
        "activo": True
    }

print(usuario)
````

````python
# 2. Acceso a elementos (por clave): usuario['username']
print(f"Usuario: {usuario['username']}")
````

````python
# 3. Agregar o actualizar elementos
usuario["rol"] = "Admin"      # Agrega nueva llave, si no existe la agrega
usuario["activo"] = False     # Actualiza llave existente

print(f"{usuario}")
````

### Acceso por medio de método *get*

````python

# 4. Acceso seguro con .get() 
# (Evita que el programa falle si la llave no existe), devolviendo el segundo parámetro cpor default
ultimo_login = usuario.get("last_login", "Nunca")
print(f"Último acceso: {ultimo_login}")
````

## Métodos Esenciales de Diccionarios

#### Para manipular la información almacenada en un `dict`, Python nos ofrece herramientas para extraer, limpiar o actualizar datos sin necesidad de acceder manualmente a cada clave.

| Método | Descripción |
| :--- | :--- |
| `.get(clave, defecto)` | Obtiene el valor de una clave. Si no existe, devuelve el valor por defecto (evita errores). |
| `.keys()` | Devuelve una lista (vista) con todas las **llaves** del diccionario. |
| `.values()` | Devuelve una lista (vista) con todos los **valores** del diccionario. |
| `.items()` | Devuelve pares de **(llave, valor)**. 
| `.update(otro_dict)` | Actualiza el diccionario con los pares clave-valor de otro diccionario. |
| `.pop(clave)` | Elimina una clave específica y devuelve su valor. |
| `.clear()` | Vacía por completo el diccionario. |

````python
servidor: dict[str, str] = {
    "id": "SRV-001",
    "status": "Online",
    "region": "MX-centro",
    "ip": "192.168.1.5"
}
````

````python

# 1. Uso de .keys()
# Útil para validar si las etiquetas obligatorias están presentes
etiquetas: list = servidor.keys()
print(f"Campos encontrados en el servidor: {etiquetas}")

if "status" in servidor.keys():
    print("Validación Exitosa: El campo 'status' existe.")
````

````python

# 2. Uso de .values()
# Útil para verificar si un valor prohibido o específico existe en el sistema
datos: list = servidor.values()
print(f"Valores almacenados: {datos}")

if "Online" in servidor.values():
    print("El servidor está disponible para pruebas.")
````

### Convertir las claves y valores a listas

````python
servidor: dict[str, str] = {
    "id": "SRV-001",
    "status": "Online",
    "region": "MX-centro",
    "ip": "192.168.1.5"
}

claves: list = list(servidor.keys())
valores: list = list(servidor.values())
print(f"las claves son: {claves}")
print(f"los valores son: {valores}")
````

# <span style="color:#0485CF"> Estructura de Datos: Tuplas (`tuple`)</span>

#### Las tuplas son colecciones de elementos similares a las listas, pero con una diferencia fundamental: son **inmutables**.

* ####  **Inmutabilidad**: Una vez creada una tupla, no puedes agregar, eliminar ni modificar sus elementos.
* ####  **Sintaxis**: Se definen utilizando paréntesis `()` 
* ####  **Rendimiento**: Son más rápidas que las listas y consumen menos memoria.
* ####  **Seguridad**: Se usan para proteger datos que no deben ser alterados por accidente (configuraciones, coordenadas, constantes).

> **Tip:** Si tus datos no van a cambiar, usa una **Tupla**. Si necesitas agregar o quitar elementos, usa una **Lista**.

## Métodos Esenciales de Tuplas

Debido a su naturaleza inmutable, las tuplas solo poseen dos métodos integrados. Estos son sumamente rápidos y eficientes para la lectura de datos.

| Método | Descripción | Caso de uso en Sistemas |
| :--- | :--- | :--- |
| `.count(valor)` | Devuelve el número de veces que un valor aparece en la tupla. | Contar cuántas veces falló un servicio en un log fijo. |
| `.index(valor)` | Devuelve el índice (posición) de la primera aparición del valor. | Encontrar la posición de una coordenada específica. |

<br>
<br>
> **Nota:** Si intentas usar `.index()` con un valor que no existe, Python lanzará un error (`ValueError`).

### 1. Definición de una tupla.
### Ejemplo: Datos de conexión que no deben cambiar


````python
# 1. Definición de una tupla (Datos de conexión que no deben cambiar)
config_db: tuple[str, int] = ("192.168.1.100", 5432)

print(f"Servidor: {config_db[0]}")
print(f"Puerto: {config_db[1]}")
````

### 2. Intentar modificar una tupla (daría error)


````python
# 2. Intentar modificar una tupla (Esto daría error)
config_db[1] = 8080  
````

### 3. Desempaquetado de Tuplas (Un truco muy usado en Python)


````python
# 3. Desempaquetado de Tuplas (Un truco muy usado en Python)
ip:int
puerto: str
ip, puerto = config_db
print(f"IP extraída: {ip}")
print(f"Puerto extraído: {puerto}")
````

### 4. Métodos disponibles (Solo los de búsqueda, no de modificación)


````python

# 4. Métodos disponibles (Solo los de búsqueda, no de modificación)
colores_rgb: tuple = (255, 0, 0, 255)
print(f"El valor 255 aparece: {colores_rgb.count(255)} veces")
print(f"El valor 0 está en el índice: {colores_rgb.index(0)}")
````
